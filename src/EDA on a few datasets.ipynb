```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from preprocessing import read_ts, stretch_interpolate, TargetEncoder, stretch_interpolate_matrix, sample_rows
```

```python
paths = os.listdir("data/")
paths
```

```python
def plot(dataset_name):
    X, y = read_ts(f'data/{dataset_name}/{dataset_name}_TRAIN.arff')  
    X, y = sample_rows(X, y, n=200)
    X = stretch_interpolate_matrix(X, 100)
    color = TargetEncoder(y).get_0_1_scaled()
    plt.figure(figsize=(15, 15))
    for i in range(len(X)):
        plt.plot(X[i, :], c=plt.cm.rainbow(color[i]))
```

```python
dataset_name = 'Beef'
plot(dataset_name)
```

```python
dataset_name = 'ECG200'
plot(dataset_name)
```

```python
dataset_name = 'Car'
plot(dataset_name)
```

```python
dataset_name = 'CricketX'
plot(dataset_name)
```

```python
dataset_name = 'SharePriceIncrease'
plot(dataset_name)
```

```python
dataset_name = 'Yoga'
plot(dataset_name)
```
